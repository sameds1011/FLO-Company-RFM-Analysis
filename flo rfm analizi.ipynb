```python
# Rfm metriklerinin hesaplanması
# Veri setindeki en son alışverişin yapıldıgı tarihten 2 gün sonrası analiz tarihi 21.05.30
df["last_order_date"].max()
```

```python
analysis_date = dt.datetime(2021,6,1)#Bugunun tarihi ayarlanması
```

```python
# customer_id,recency,frequnecy,monetary değerlerinin yer aldıgı yeni bir rfm dataframe
rfm=pd.DataFrame()
rfm["custormer_id"] =df["master_id"]#Müşteri verilerinin tutarlılıgını artırmak CRM sistemleriyle entegre olurken veri tutarlılığını sağlar
```

```python
rfm.head()
```

```python
rfm["recency"]=(analysis_date-df['last_order_date']).dt.days #Bugünden en son alışveriş yapılan tarihin gününü çıkar.
#Bugunun tanımı analysis date dir #Recency Enson ne zaman alışveriş yaptıgı 
```

```python
rfm.head()# ilgili müşterilerin örnek 0 numara index 1 en son 95 gün önce alışveriş yapmış 1 numara enson  105 gün önce alışveriş yapmış
```

```python
rfm["frequency"]=df["order_num_total"] #Frequency=sıklık Müşteri alışveriş Sıklıgını Belirler.
rfm.head()
# index 0 da recency de görüldügü üzere 95 gün önce alış veriş yapmış en son.
#recency_score müşteriye atanan degerdir
#En degerli müşteri 1 numaralı 105 gün önce 21 kez alış veriş yapan En degersiz müşteri 2 numara 186 günde 2 kez alışveriş yapmış
```

```python
rfm["monetary"]=df["customer_value_total"]# Parasal deger bilgileri için monetary eşitlenir
rfm.head()
# Müşterinin şirketin gelirine olan katkısını anlamak müşterinin geçmişteki alışverişlerinden elde ettiği toplam parasal değeri
# monetary müşterinin şirkete harcadıgı miktar yani yaptıgı katkıdır.
#En sık alışveriş yapan müşteri frequency 21 kez. Toplam harcaması da monetary 2013.55 birim. Bu müşteri en değerli müşteridir 1 numara index 1 f43
```

```python
# RF ve RFM skorlarının hesaplanması
# Recency Frequency Monetary metriklerini qcut yardımı ile 1-5 arasında skorlara çevrilmesi
# ve bu skorları recency_skore,frequecny_skore monetary_score olarak kaydedilir
rfm["recency_score"]=pd.qcut(rfm['recency'],5,labels=[5,4,3,2,1]) #recency degerlerini 5 parçaya böl labelsde 5 4 3 2 1 diye puanlar qcut kullanır
```

```python
rfm["recency_score"]=pd.qcut(rfm['recency'],5, labels=[5,4,3,2,1])
```

```python
rfm.head() # 3 ortalama puandır ne çok degerli ne çok degersizdir  2 degersizdir
```

```python
pd.crosstab(rfm["recency"],rfm["recency_score"]).head(20)
#2 gün önce gelen 80  müşteri var mış 5 e koymuş degerlendirmesi iyi diye teşekkür mesajı atılabilir veya özel kampanyalar oluşturulabilir.
#8 gün önce gelene tekrardan işlem yapabilir gibi
#pd.crosstab İki değişken arasında karşılaştırarak bir tablo oluşturur.her bir recency değeri için kaç tane recency_score olduğunu gösterir.
# recency_score 4 3 2 1 degerinin olmaması öyle müşterinin olmaması veya müşteri segmentinin hiç oluşmadıgı düşünülebilir.
```

```python
rfm["frequency_score"]=pd.qcut(rfm['frequency'].rank(method="first"),5,labels=[1,2,3,4,5]) #Sıklık puanını hesaplar
#pd.qcut verileri 5 eşit parçaya böler 1 ile 5 arasında her parçaya etiket atar 
```

```python
pd.crosstab(rfm["frequency"],rfm["frequency_score"]).head(20)
# 2.00 yani 2 kez toplamda alış verişe gelen müşteriler 3989 müşteri 1 puan almış 728 müşteri 2 puan almış çok düşük alışveriş yapan büyük bir müşteri grubu

# En yüksek sıklık 7.00 978 müşteri 5 puan almış.En az sıklık gösteren alışverişte müşteriler 2.00 3989 müşteri 1 ouan almış nadiren alışveriş yapan
#
```

```python
rfm["monetary_score"]= pd.qcut(rfm['monetary'], 5, labels=[1, 2,3, 4, 5])
```

```python
pd.crosstab(rfm["monetary"],rfm["monetary_score"]).head(20)
# 44.98 birim harcayan 2 müşteri 1 puan almış
#Düşük harcama yapan müşterilere yönelik kampanyalar oluşturmak ve yüksek harcama yapan müşterileri ödüllendirmek öneririm.
```

```python
# recency_score frequency score tek bir degişken olarak ifade edelim ve rf_score olarak kaydedelim.
rfm["RF_SCORE"]=(rfm['recency_score'].astype(str)+rfm['frequency_score'].astype(str))
```

```python
rfm.head()
```

```python
# RF Skorlarının segment olarak tanımlanması rfm skorlarının daha açıklanabilir olması için
# Bu yöntemde müşteri bağlılığını artırmanın yanı sıra müşteri kaybını önlemeye ve işletmenin gelirlerini artırmaya yardımcı olur.
seg_map = {
    r'[1-2][1-2]': 'hibernating',# uyuyan müşteriler: Bu segmentte yer alan müşteriler hem düşük sıklıkla  hem de düşük harcamalarla tanımlanır. Bu müşteriler uzun zamandır aktif değil ve az harcama yapmışlar. 
    r'[1-2][3-4]': 'at_Risk',# Risk altında müşteriler:Yüksek sıklıkta alışveriş yapan uzun süre önce alışveriş yapmış müşteriler Şirket bu müşterileri kaybedebilir.
    r'[1-2]5': 'cant_loose', # kaybedilemeyecek müşteriler:Bu müşteriler sık alışveriş yapmış ve yüksek miktarda harcama yapmışlar. ancak yakın zamanda alışveriş yapmamışlar. Onları tekrar kazanmak kritik olabilir.
    r'3[1-2]': 'about_to_sleep',#Uykuya geçen müşteriler:Uzun süre önce aktif olan orta seviye ve sıklıkta harcama yapan müşteriler.
    r'41': 'promising', # vaat eden müşteriler: Bu yeni müşteriler, kısa süre önce bir alışveriş yapmış ancak henüz çok fazla alışveriş yapmamışlar
    r'51': 'new_customers', # yeni müşteriler: Yakın zamanda alışveriş yapmış olan yeni müşterileri temsil eder. Henüz sık alışveriş yapmamışlar ama onları bağlı müşteriler haline getirme potansiyeli vardır.
    r'[4-5][2-3]': 'potential_loyalists',# potansiyel sadık müşteriler: Bu müşteriler orta sıklıkta ve orta harcama yapmışlar. Sadık birer müşteri olma potansiyeline sahipler.
    r'5[4-5]': 'champions'  # şampiyon müşteriler: Bu müşteriler hem sık alışveriş yapmış, hem de yüksek miktarda harcama yapmışlardır. En değerli ve sadık müşteri grubunu oluştururlar.
}

# Segment atama işlemi (girintileme ve atama hatası düzeltildi)
rfm['segment'] = rfm['RF_SCORE'].replace(seg_map, regex=True)

# Sonuçları görmek için ilk 5 satırı görüntüleme
rfm.head()
```

```python
rfm[["segment","recency","frequency","monetary"]].groupby("segment").agg(["mean","count"])
#Segmentlerin ortalamaları
```

```python
# Hedef segmentlerdeki müşteri ID'lerini seçiyoruz (champions ve loyal_customers)
target_segments_customer_ids = rfm[rfm['segment'].isin(["champions", "loyal_customers"])]["customer_id"]

# Hedef müşteri segmentlerine ait ve KADIN kategorisine ilgi duyan müşterileri seçiyoruz
cust_ids = df[(df["master_id"].isin(target_segments_customer_ids)) &
              (df["interested_in_categories_12"].str.contains("KADIN"))]["master_id"]

# Müşteri ID'lerini CSV dosyasına kaydetme
cust_ids.to_csv("yeni_marka_hedef_müşteri_id.csv", index=False)

# Kaç müşterinin olduğunu görmek için
print("Toplam müşteri sayısı:", cust_ids.shape[0])

```
